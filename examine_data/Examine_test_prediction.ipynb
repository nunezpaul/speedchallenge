```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams['font.size'] = 18
MAX_LEN = 10797
```

```python
BUCKET_SIZE = 3

y_pred_3 = pd.read_csv('../data/predictions/test/test_bucket_size_3_10_buckets.txt')
y_pred_3 = np.argmax(a=y_pred_3.values, axis=-1)[:MAX_LEN]
y_pred_speed_3 = (y_pred_3 + 0.5) * BUCKET_SIZE
```

```python
BUCKET_SIZE = 2

y_pred_2 = pd.read_csv('../data/predictions/test/test_bucket_size_2_15_buckets.txt')
y_pred_2 = np.argmax(a=y_pred_2.values, axis=-1)[:MAX_LEN]
y_pred_speed_2 = (y_pred_2 + 0.5) * BUCKET_SIZE
```

```python
BUCKET_SIZE = 1

y_pred_1 = pd.read_csv('../data/predictions/test/test_bucket_size_1_29_buckets.txt')
y_pred_1 = np.argmax(a=y_pred_1.values, axis=-1)[:MAX_LEN]
y_pred_speed_1 = (y_pred_1 + 0.5) * BUCKET_SIZE
```

```python
y_pred_speed_avg = np.mean([y_pred_speed_1, y_pred_speed_2, y_pred_speed_3], axis = 0)
```

```python
plt.plot(y_pred_speed_1)
plt.plot(y_pred_speed_2)
plt.plot(y_pred_speed_3)
plt.plot(y_pred_speed_avg, linewidth=4)

plt.xlabel('Frame')
plt.ylabel('Speed (m/s)')
plt.title('Speed Prediction of Test File')

plt.show()
```
